## Visualization & Explanatory Analysis

### Pertanyaan 1: 

```python
# Pada musim apa dalam 2 tahun terjadi puncak aktivitas penyewaan sepeda?
```

```python
# Membuat visualisasi data dari pertanyaan pertama
byseason_df = day_df.groupby(by="season").dteday.nunique().reset_index()
byseason_df.rename(columns={
    "dteday" : "season_count"
}, inplace=True)

colors = ["#D3D3D3", "#D3D3D3", "#72BCD4", "#D3D3D3"]

plt.figure(figsize=(12, 5))

sns.barplot(
    y= "season_count",
    x="season",
    data=byseason_df.sort_values(by="season_count", ascending=False),
    palette=colors
)
plt.title("Season Terbanyak pada Penyewaan Sepeda", loc="center", fontsize=15)
plt.ylabel("Jumlah Penyewa")
```

### Pertanyaan 2: 

```python
# Pada 10 hari terakhir, penyewa sepeda paling banyak berasal dari penyewa casual atau penyewa registered?
```

```python
# Membuat visualisasi data dari pertanyaan kedua
casual_df = day_df.groupby(by="dteday").casual.sum().sort_values(ascending=False).reset_index()
member_df = day_df.groupby(by="dteday").registered.sum().sort_values(ascending=False).reset_index()
jumlah_penyewa_df = day_df.casual.sum()
fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(24, 6))
 
colors = ["#72BCD4", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3", "#D3D3D3"]

sns.barplot(x="casual", y="dteday", data=casual_df.head(10), palette=colors, ax=ax[0])
ax[0].set_ylabel(None)
ax[0].set_xlabel(None)
ax[0].set_title("Penyewa General", loc="center", fontsize=15)
ax[0].tick_params(axis ='y', labelsize=12)

sns.barplot(x="registered", y="dteday", data=member_df.head(10), palette=colors, ax=ax[1])
ax[1].set_ylabel(None)
ax[1].set_xlabel(None)
ax[1].invert_xaxis()
ax[1].yaxis.set_label_position("right")
ax[1].yaxis.tick_right()
ax[1].set_title("Penyewa Member atau Terdaftar", loc="center", fontsize=15)
ax[1].tick_params(axis='y', labelsize=12)
 
plt.suptitle("Perbandingan Penyewa Antara General dan Member", fontsize=20)
plt.show()

```
