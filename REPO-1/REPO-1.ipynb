```python
import numpy as np
import pandas as pd
```

```python
# Membaca dataset dan menampilkan nya
customers_df = pd.read_csv('./customers.csv')

customers_df.head()
```

```python
# Menampilkan info dari dataset
customers_df.info()
```

```python
# Mengecek apakah terdapat data yang sama atau yang kosong

customers_df.isna().sum()
```

```python
print("Jumlah Duplikasi Data : ", customers_df.duplicated().sum())
```

```python
# Menampilakn deskripsi dari dataset

customers_df.describe()
```

```python
# Menghapus duplikasi data

customers_df.drop_duplicates(inplace=True)
```

```python
print("Jumlah Duplikasi Data : ", customers_df.duplicated().sum())
```

```python
# Menangani Missing Value

customers_df[customers_df.gender.isna()]
```

```python
# Menggunakan metode imputation untuk mengganti nilai dari missing value

customers_df.gender.value_counts()
```

```python
# Menggunakan data prefer not to say sebagai data missing dan mengganti nya menggunakan method fillna

customers_df.fillna(value="Prefer not to say", inplace=True)
```

```python
customers_df.isna().sum()
```

```python
# Menangani inacurate Value

customers_df[customers_df.age == customers_df.age.max()]
```

```python
# Menambahkan method replace pada data yang termasuk inaccurate

customers_df.age.replace(customers_df.age.max(), 70, inplace=True)
```

```python
customers_df[customers_df.age == customers_df.age.max()]
```
